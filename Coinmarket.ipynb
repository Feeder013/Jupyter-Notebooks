```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# **COINMARKETCAP WEBSCRAPING**
  
This project utilizes Python to scrape and extract data on the top 500 cryptocurrencies from CoinMarketCap on a enabling comparative analysis of trends and patterns in cryptocurrency performance and the development of predictive models for future market trends.

```python
# Enter the coinmarketcap API
url = 'https://coinmarketcap.com/'
```

```python
# To extract information from the first 50 pages
coin_list = []
for plist in range(1,51):
  coin_list.append('https://coinmarketcap.com/?page='+ str(plist))

len(coin_list)
```

```python
#To prevent blocking use user agent
User_Agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'

headers = {'User-Agent': User_Agent}

#the following parameters allows the code to extract more than the top 10 rows which is ideally scraped without it
params = {
    'start': '1',
    'limit': '100',
    'sort': 'market_cap',
    'sort_dir': 'desc',
    'cryptoType': 'all',
    'tagType': 'all',
    'aux': 'ath,atl,high24h,low24h,num_market_pairs,cmc_rank,date_added,tags,platform,max_supply,circulating_supply'
}

```

```python
#definition a function to a bs4 to the html each time its ran
def get_url(url):
    result = requests.get(url, headers=headers, params=params)
    soup = BeautifulSoup(result.text, 'html.parser')
    return soup
#creating a variable for each column that is to be desired
Name = []
Price = []
onehour  = []
twentyfourhour  = []
sevendayz = []
Market_Cap = []
Volume_24hour = []
Circulating_Supply = []

#to select each pages from begining to end out of the 50 page list
for results in coin_list[::]:
  soup = get_url(results)
#the following are the lines the desired codes are found during inspection
#<p class="sc-65e7f566-0 iPbTJf coin-item-name">Bitcoin</p>
#<span class="sc-65e7f566-0 WXGwg base-text" data-test="text-cdp-price-display">$91,212.46</span>
#<div class="sc-b3fc6b7-0 dzgUIj"><span>$1.39</span></div>
#<span class="sc-a59753b0-0 cmnujh"><span class="icon-Caret-up" style="width:12px;height:18px;display:inline-block"></span>0.16%</span>
#<span data-nosnippet="true" class="sc-11478e5d-1 jfwGHx">$1,805,830,092,894</span>
#<p font-size="1" class="sc-71024e3e-0 bbHOdE font_weight_500" color="text" data-sensors-click="true">$69,701,051,751</p>
#<p font-size="1" font-weight="medium" color="text" data-sensors-click="true" class="sc-71024e3e-0 hhmVNu">19,782,884 BTC</p>
  for name_list in soup.find_all('p', class_='sc-65e7f566-0 iPbTJf coin-item-name'):
    try:
      Name.append(name_list.text.replace('\n',''))
    except:
      Name.append('np.nan')
  for price_list in soup.find_all('div', class_='sc-b3fc6b7-0 dzgUIj'):
      try:
        Price.append(price_list.text.replace('\n',''))
      except:
        Price.append('np.nan')
  for onehour_list in soup.find_all('span', class_ ='sc-a59753b0-0 cmnujh'):
      try:
        onehour.append(onehour_list.text.replace('\n',''))
      except:
        onehour.append('np.nan')
  for twentyfourhour_list in soup.find_all('span', class_ ='sc-a59753b0-0 cmnujh'):
      try:
        twentyfourhour.append(twentyfourhour_list.text.replace('\n',''))
      except:
        twentyfourhour.append('np.nan')
  for sevendayz_list in soup.find_all('span', class_ ='sc-a59753b0-0 cmnujh'):
      try:
        sevendayz.append(sevendayz_list.text.replace('\n',''))
      except:
        sevendayz.append('np.nan')
  for market_cap_list in soup.find_all('span', class_='sc-11478e5d-1 jfwGHx'):
      try:
        Market_Cap.append(market_cap_list.text.replace('\n',''))
      except:
        Market_Cap.append('np.nan')
  for volumne24_list in soup.find_all('p', class_ = 'sc-71024e3e-0 bbHOdE font_weight_500', attrs={'data-sensors-click':'true'}):
      try:
        Volume_24hour.append(volumne24_list.text.replace('\n',''))
      except:
        Volume_24hour.append('np.nan')
  for circulating_supply_list in soup.find_all('p', class_ = 'sc-71024e3e-0 hhmVNu'):
      try:
        Circulating_Supply.append(circulating_supply_list.text.replace('\n',''))
      except:
        Circulating_Supply.append('np.nan')
```

```python
#extracting column names separately for the dataframe
columnz = soup.find_all('p', class_='sc-71024e3e-0 fSsxNG', attrs={'data-sensors-click':'true'})
columnz

column = [column.text for column in columnz]
column
df = pd.DataFrame(columns=column)
df
#test run by adding names to the dataframe
df['Name'] = Name
df
```

```python
#importing numpy array to fill up the rest of the table as the scraped data are less and some more than the length of the dataframe
import numpy as np
min_len = min(len(df), len(onehour))

df['1h %'] = onehour[:min_len]
pri_len = min(len(df), len(Price))
df['Price'] = Price[:pri_len]
T_len = min(len(df), len(twentyfourhour))
df['24h %'] = twentyfourhour[:T_len]
seven_len = min(len(df), len(sevendayz))
df['7d %'] = sevendayz[:seven_len]
market_len = min(len(df), len(Market_Cap))
# Extend Market_Cap with NaN values to match DataFrame length
Market_Cap.extend([np.nan] * (len(df) - market_len))
df['Market Cap'] = Market_Cap[:len(df)]

volume_len = min(len(df), len(Volume_24hour))
Volume_24hour.extend([np.nan] * (len(df) - volume_len))
df['Volume(24h)'] = Volume_24hour[:len(df)]

circulating_len = min(len(df), len(Circulating_Supply))
Circulating_Supply.extend([np.nan] * (len(df) - circulating_len))
df['Circulating Supply'] = Circulating_Supply[:len(df)]
df
```

```python
coinmarketcap = df
coinmarketcap.to_csv(r'/coinmarketcapscrapedON16-11-24.csv')
```
